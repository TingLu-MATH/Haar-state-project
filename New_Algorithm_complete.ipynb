```python
from sympy import *
import numpy as np
import os
import sys
### Path to the folder where the script Modules.py is stored.
sys.path.append('/Users/tinglu/Desktop/Computational_code/')
import Modules as md
q=symbols('q')
```

```python
### the order of standard monomials
m=5
```

```python
[dict_1,dict_2]=md.generate_dict(m,'l')
### Path to the folder where the Haar states of order m-1 standard monomials are stored
pathA='/Users/tinglu/Desktop/Computational_code/Results/Order_4/'
### Path to the folder where you want to store the Haar stated of order m standard monomials
pathB='/Users/tinglu/Desktop/Computational_code/Results/Order_5/'
```

```python
### i = number of afh segment 
for i in range(1,m):
    for j in range(m-i):
        for k in range(m-i-j):
            s='ceg'+md.IndtoStMo([0,i,0,j,k,m-i-j-k-1],'l')
            l1=md.term_expansion(s,0,0,'l',dict_1,dict_2,0)
            H=0
            for l in range(len(dict_1)):
                if l1[l]!=0 and l!=dict_2[md.IndtoStMo([0,i,0,j,k,m-i-j-k],'l')]:
                    f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[l],'l'))).replace(' ','')+'.txt','r')
                    r=eval(str(f.read()))
                    f.close()
                    H=simplify(H+l1[l]*r)
            d=factor(H/(1-q**(2*i)))
            f1=open(pathB+str([0,i,0,j,k,m-i-j-k]).replace(' ','')+'.txt','w')
            f1.write(str(d))
            f1.close()
            f1=open(pathB+str([0,0,i,k,j,m-i-j-k]).replace(' ','')+'.txt','w')
            f1.write(str(d))
            f1.close()
    for j in range(m-i+1):
        t=md.generate_linear_relation([0,i-1,0,m-i-j,j+1,0],'bdk',m,dict_1,dict_2)
        H=0
        for l in range(len(dict_1)):
            if l!=dict_2[md.IndtoStMo([0,i,0,m-i-j,j,0],'l')] and t[l]!=0:
                f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[l],'l'))).replace(' ','')+'.txt','r')
                r=eval(str(f.read()))
                f.close()
                H=simplify(H+t[l]*r)
        d=factor(-H/t[dict_2[md.IndtoStMo([0,i,0,m-i-j,j,0],'l')]])
        f=open(pathB+str([0,i,0,m-i-j,j,0]).replace(' ','')+'.txt','w')
        f.write(str(d))
        f.close() 
        f=open(pathB+str([0,0,i,j,m-i-j,0]).replace(' ','')+'.txt','w')
        f.write(str(d))
        f.close() 
        
#### i = m

t=md.generate_linear_relation([0,m-1,0,0,1,0],'bdk',m,dict_1,dict_2)
H=0
for l in range(len(dict_1)):
    if l!=dict_2[md.IndtoStMo([0,m,0,0,0,0],'l')] and t[l]!=0:
        f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[l],'l'))).replace(' ','')+'.txt','r')
        r=eval(str(f.read()))
        f.close()
        H=simplify(H+t[l]*r)
d=factor(-H/t[dict_2[md.IndtoStMo([0,m,0,0,0,0],'l')]])
f=open(pathB+str([0,m,0,0,0,0]).replace(' ','')+'.txt','w')
f.write(str(d))
f.close()       
f=open(pathB+str([0,0,m,0,0,0]).replace(' ','')+'.txt','w')
f.write(str(d))
f.close() 
```

```python
### compute the Haar stat of monomials with only one aek segment and no other high-complexity segments
#j = number of ceg segments (the Haar state of aek(ceg)^(m-1) is known)
for j in range(m-1):
    # k = number of cdh segments
    for k in range(m-j):
        f=open(pathA+str([0,0,0,m-1-j-k,k,j]).replace(' ','')+'.txt','r')
        r0=eval(str(f.read()))
        f.close()
        f=open(pathB+str([0,1,0,m-1-j-k,k,j]).replace(' ','')+'.txt','r')
        r1=eval(str(f.read()))
        f.close()
        f=open(pathB+str([0,0,1,m-1-j-k,k,j]).replace(' ','')+'.txt','r')
        r2=eval(str(f.read()))
        f.close()
        f=open(pathB+str([0,0,0,m-j-k,k,j]).replace(' ','')+'.txt','r')
        r3=eval(str(f.read()))
        f.close()
        f=open(pathB+str([0,0,0,m-1-j-k,k+1,j]).replace(' ','')+'.txt','r')
        r4=eval(str(f.read()))
        f.close()
        f=open(pathB+str([0,0,0,m-1-j-k,k,j+1]).replace(' ','')+'.txt','r')
        r5=eval(str(f.read()))
        f.close()
        r=factor(r0+q*r1+q*r2-q**2*r3-q**2*r4+q**3*r5)
        f=open(pathB+str([1,0,0,m-1-j-k,k,j]).replace(' ','')+'.txt','w')
        f.write(str(r))
        f.close()
```

```python
### general form

# i = number of high-complexity segments
for i in range(2,m):
    #1. compute the Haar state of monomials with ceg segments and without aek segments
    # u = number of afh segments:
    for u in range(1,int(i/2)+1):
        # j = number of ceg segments
        for j in range(1,m-i+1):
            # k = number of cdh segments
            for k in range(0,m-i-j+1):
                l2=md.term_expansion(md.IndtoStMo([0,u,i-u,m-i-j-k,k,j],'l'),0,0,'l',dict_1,dict_2,0)
                H=0
                for t in range(len(dict_1)):
                    if l2[t]!=0:
                        f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[t],'l'))).replace(' ','')+'.txt','r')
                        r=eval(str(f.read()))
                        f.close()
                        H=simplify(H+l2[t]*r)
                H=factor(H)
                f=open(pathB+str([0,u,i-u,m-i-j-k,k,j]).replace(' ','')+'.txt','w')
                f.write(str(H))
                f.close()
                f=open(pathB+str([0,i-u,u,k,m-i-j-k,j]).replace(' ','')+'.txt','w')
                f.write(str(H))
                f.close()
    #2. compute the Haar state of monomials without ceg segments and without aek segments
    # u = number of afh segments:
    for u in range(1,int(i/2)+1):
        # k = number of cdh segments
        for k in range(0,m-i+1):
            l3=md.generate_linear_relation([0,u,i-u-1,m-i-k+1,k,0],'afh',m,dict_1,dict_2)
            H=0
            for l in range(len(dict_1)):
                if l!=dict_2[md.IndtoStMo([0,u,i-u,m-i-k,k,0],'l')] and l3[l]!=0:
                    f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[l],'l'))).replace(' ','')+'.txt','r')
                    r=eval(str(f.read()))
                    f.close()
                    H=simplify(H+l3[l]*r)
            d=factor(-H/l3[dict_2[md.IndtoStMo([0,u,i-u,m-i-k,k,0],'l')]])
            f=open(pathB+str([0,u,i-u,m-i-k,k,0]).replace(' ','')+'.txt','w')
            f.write(str(d))
            f.close() 
            f=open(pathB+str([0,i-u,u,k,m-i-k,0]).replace(' ','')+'.txt','w')
            f.write(str(d))
            f.close()
    #3. compute the Haar state of monomials with aek segments
    # v = number of aek segments
    for v in range(1,i+1):
        # w = number of afh segments
        for w in range(i-v+1):
            # j = number of ceg segments
            for j in range(m-i+1):
                # k = number of cdh segments
                for k in range(m-i-j+1):
                    f=open(pathA+str([v-1,w,i-v-w,m-i-j-k,k,j]).replace(' ','')+'.txt','r')
                    r0=eval(str(f.read()))
                    f.close()
                    f=open(pathB+str([v-1,w+1,i-v-w,m-i-j-k,k,j]).replace(' ','')+'.txt','r')
                    r1=eval(str(f.read()))
                    f.close()
                    s2=md.IndtoStMo([v-1,0,0,0,0,0],'l')+'bdk'+md.IndtoStMo([0,w,i-v-w,m-i-j-k,k,j],'l')
                    l2=md.term_expansion(s2,0,0,'l',dict_1,dict_2,0)
                    r2=0
                    for t in range(len(dict_1)):
                        if l2[t]!=0:
                            f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[t],'l'))).replace(' ','')+'.txt','r')
                            H=eval(str(f.read()))
                            f.close()
                            r2=simplify(r2+l2[t]*H)
                    s3=md.IndtoStMo([v-1,0,0,0,0,0],'l')+'bfg'+md.IndtoStMo([0,w,i-v-w,m-i-j-k,k,j],'l')
                    l3=md.term_expansion(s3,0,0,'l',dict_1,dict_2,0)
                    r3=0
                    for t in range(len(dict_1)):
                        if l3[t]!=0:
                            f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[t],'l'))).replace(' ','')+'.txt','r')
                            H=eval(str(f.read()))
                            f.close()
                            r3=simplify(r3+l3[t]*H)
                    s4=md.IndtoStMo([v-1,0,0,0,0,0],'l')+'cdh'+md.IndtoStMo([0,w,i-v-w,m-i-j-k,k,j],'l')
                    l4=md.term_expansion(s4,0,0,'l',dict_1,dict_2,0)
                    r4=0
                    for t in range(len(dict_1)):
                        if l4[t]!=0:
                            f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[t],'l'))).replace(' ','')+'.txt','r')
                            H=eval(str(f.read()))
                            f.close()
                            r4=simplify(r4+l4[t]*H)
                    s5=md.IndtoStMo([v-1,0,0,0,0,0],'l')+'ceg'+md.IndtoStMo([0,w,i-v-w,m-i-j-k,k,j],'l')
                    l5=md.term_expansion(s5,0,0,'l',dict_1,dict_2,0)
                    r5=0
                    for t in range(len(dict_1)):
                        if l5[t]!=0:
                            f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[t],'l'))).replace(' ','')+'.txt','r')
                            H=eval(str(f.read()))
                            f.close()
                            r5=simplify(r5+l5[t]*H)
                    r=factor(r0+q*r1+q*r2-q**2*r3-q**2*r4+q**3*r5)
                    f=open(pathB+str([v,w,i-v-w,m-i-j-k,k,j]).replace(' ','')+'.txt','w')
                    f.write(str(r))
                    f.close()
```

```python
### compute the Haar state of monomials with no low-complexity segments
# 1. no aek segments
for u in range(1,int(m/2)+1):
    l3=md.generate_linear_relation([0,u,m-u-1,1,0,0],'afh',m,dict_1,dict_2)
    H=0
    for l in range(len(dict_1)):
        if l!=dict_2[md.IndtoStMo([0,u,m-u,0,0,0],'l')] and l3[l]!=0:
            f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[l],'l'))).replace(' ','')+'.txt','r')
            r=eval(str(f.read()))
            f.close()
            H=simplify(H+l3[l]*r)
    d=factor(-H/l3[dict_2[md.IndtoStMo([0,u,m-u,0,0,0],'l')]])
    f=open(pathB+str([0,u,m-u,0,0,0]).replace(' ','')+'.txt','w')
    f.write(str(d))
    f.close() 
    f=open(pathB+str([0,m-u,u,0,0,0]).replace(' ','')+'.txt','w')
    f.write(str(d))
    f.close()
```

```python
# 2. with aek segments
# v = number of aek segments
for v in range(1,m+1):
        # w = number of afh segments
        for w in range(int((m-v)/2)+1):
            f=open(pathA+str([v-1,w,m-v-w,0,0,0]).replace(' ','')+'.txt','r')
            r0=eval(str(f.read()))
            f.close()
            f=open(pathB+str([v-1,w+1,m-v-w,0,0,0]).replace(' ','')+'.txt','r')
            r1=eval(str(f.read()))
            f.close()
            s2=md.IndtoStMo([v-1,0,0,0,0,0],'l')+'bdk'+md.IndtoStMo([0,w,m-v-w,0,0,0],'l')
            l2=md.term_expansion(s2,0,0,'l',dict_1,dict_2,0)
            r2=0
            for t in range(len(dict_1)):
                if l2[t]!=0:
                    f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[t],'l'))).replace(' ','')+'.txt','r')
                    H=eval(str(f.read()))
                    f.close()
                    r2=simplify(r2+l2[t]*H)
            s3=md.IndtoStMo([v-1,0,0,0,0,0],'l')+'bfg'+md.IndtoStMo([0,w,m-v-w,0,0,0],'l')
            l3=md.term_expansion(s3,0,0,'l',dict_1,dict_2,0)
            r3=0
            for t in range(len(dict_1)):
                if l3[t]!=0:
                    f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[t],'l'))).replace(' ','')+'.txt','r')
                    H=eval(str(f.read()))
                    f.close()
                    r3=simplify(r3+l3[t]*H)
            s4=md.IndtoStMo([v-1,0,0,0,0,0],'l')+'cdh'+md.IndtoStMo([0,w,m-v-w,0,0,0],'l')
            l4=md.term_expansion(s4,0,0,'l',dict_1,dict_2,0)
            r4=0
            for t in range(len(dict_1)):
                if l4[t]!=0:
                    f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[t],'l'))).replace(' ','')+'.txt','r')
                    H=eval(str(f.read()))
                    f.close()
                    r4=simplify(r4+l4[t]*H)
            s5=md.IndtoStMo([v-1,0,0,0,0,0],'l')+'ceg'+md.IndtoStMo([0,w,m-v-w,0,0,0],'l')
            l5=md.term_expansion(s5,0,0,'l',dict_1,dict_2,0)
            r5=0
            for t in range(len(dict_1)):
                if l5[t]!=0:
                    f=open(pathB+str(md.CMtoInd(md.MonToCM(dict_1[t],'l'))).replace(' ','')+'.txt','r')
                    H=eval(str(f.read()))
                    f.close()
                    r5=simplify(r5+l5[t]*H)
            r=factor(r0+q*r1+q*r2-q**2*r3-q**2*r4+q**3*r5)
            f=open(pathB+str([v,w,m-v-w,0,0,0]).replace(' ','')+'.txt','w')
            f.write(str(r))
            f.close()
            f=open(pathB+str([v,m-v-w,w,0,0,0]).replace(' ','')+'.txt','w')
            f.write(str(r))
            f.close()
```
